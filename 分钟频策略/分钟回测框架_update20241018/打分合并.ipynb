```python
import pandas as pd
import numpy as np
import os
import sys
from tqdm.notebook import tqdm
```

```python
path1 = r'/home/datamake94/database_00/min_score_period30/min_score_lgbm_new_241016_select4_23'
path2 = r'/home/datamake94/database_00/min_score_period30/min_score_lgbm_new_241016_select4_interval2'

date_list = os.listdir(path2)
date_list = sorted([x[:8] for x in date_list])

for date in tqdm(date_list[:-1]):
    file1 = pd.read_feather(os.path.join(path1, date + '.fea')).set_index('index')
    file2 = pd.read_feather(os.path.join(path2, date + '.fea')).set_index('index')
    file1.loc[1000:] = file2.loc[1000:]
    file1.reset_index().to_feather(os.path.join(path1, date + '.fea'))
```
